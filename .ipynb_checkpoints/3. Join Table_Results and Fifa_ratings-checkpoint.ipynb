```python
Table_Results = pd.read_excel('Table_results.xlsx')
Table_Results.drop(['Unnamed: 0'],axis = 1, inplace = True)
Table_Results
```

```python
# Get the team ratings from Fifa index website

year = 2020

   
webpage1 = pd.read_html(f'https://www.fifaindex.com/teams/fifa{str(year)[-2:]}/?league=308&order=desc')
Teams_Rating = pd.DataFrame(webpage1[0])
Teams_Rating.dropna(subset = ["Name"], inplace=True)
Teams_Rating.drop(Teams_Rating.columns.difference(['Name','League','ATT','MID','DEF','OVR']), 1, inplace=True)
Teams_Rating['Year'] = Year
Teams_Rating['Ranking'] = range(1,19)

column_names = ['Year', 'Ranking', 'Name', 'League', 'ATT', 'MID', 'DEF', 'OVR']
Teams_Rating = Teams_Rating.reindex(columns = column_names)

Teams_Rating = Teams_Rating.set_index('Year')

Teams_Rating
```

```python
list_unique_teams_Table_Results = Table_Results['Home_team'].unique().tolist()
list_unique_teams_Table_Results.sort()

print(f'{len(list_unique_teams_Table_Results)} teams in Table_Results:')
print('\n')
print(list_unique_teams_Table_Results)
print('\n')

teams_list_fifa = Teams_Rating['Name'].tolist()
teams_list_fifa.sort()

print(f'{len(teams_list_fifa)} teams in Fifa_index')
print('\n')
print(teams_list_fifa)
```

```python
# I manually added the Fifa corresponding name, but there must be a better option for this.

Teams_Names = pd.DataFrame({'Table_Results_Teams_Name':list_unique_teams_Table_Results})

FIFA_Teams_Names_List = [
    'Unknown_Académica',
    'Unknown_Arouca',
    'SL Benfica',
    'Boavista FC',
    'Santa Clara',
    'Unknown_Chaves',
    'Desportivo das Aves',
    'Unknown_Estoril',
    'Futebol Clube de Famalicão',
    'Unknown_Feirense',
    'Gil Vicente FC',
    'Marítimo',
    'Moreirense FC',
    'Unknown_Nacional',
    'Belenenses',
    'FC Paços de Ferreira',
    'Portimonense SC',
    'FC Porto',
    'Rio Ave FC',
    'SC Braga',  
    'Sporting CP',
    'CD Tondela',
    'Unknown_UniaoMadeira',
    'Vitória Guimarães',
    'Vitória Setúbal'
]

Teams_Names['Fifa_index_Teams_Name'] = FIFA_Teams_Names_List

Teams_Names
```

```python
Table_Results_with_FIFA = Table_Results
```

```python
# Create 2 new columns for the home and away teams with the Fifa name on it

# Home_teams_names

list_Home_team_Table_Results_with_Fifa = Table_Results_with_FIFA['Home_team'].tolist()

list_Home_team_Table_Results_with_Fifa_for_Fifa_column = []

for n in list_Home_team_Table_Results_with_Fifa:
    Team_name_Fifa = Teams_Names.loc[Teams_Names['Table_Results_Teams_Name'] == n, 'Fifa_index_Teams_Name'].iloc[0]
    list_Home_team_Table_Results_with_Fifa_for_Fifa_column.append(Team_name_Fifa)

Table_Results_with_FIFA['Home_team_Fifa'] = list_Home_team_Table_Results_with_Fifa_for_Fifa_column

# Away_teams_names

list_Away_team_Table_Results_with_Fifa = Table_Results_with_FIFA['Away_team'].tolist()

list_Away_team_Table_Results_with_Fifa_for_Fifa_column = []

for n in list_Away_team_Table_Results_with_Fifa:
    Team_name_Fifa = Teams_Names.loc[Teams_Names['Table_Results_Teams_Name'] == n, 'Fifa_index_Teams_Name'].iloc[0]
    list_Away_team_Table_Results_with_Fifa_for_Fifa_column.append(Team_name_Fifa)

Table_Results_with_FIFA['Away_team_Fifa'] = list_Away_team_Table_Results_with_Fifa_for_Fifa_column
```

```python
# Re order columns to have the team names columns together.

Table_Results_with_FIFA = Table_Results_with_FIFA[['Year','Weekday',
                                                   'Home_team','Away_team',
                                                   'Home_team_Fifa','Away_team_Fifa',
                                                   'Result','Home_score','Away_score',
                                                   '1x2','Points_Home_Team','Points_Away_Team']]
```

```python
# test to see how to add ratings

import warnings
warnings.filterwarnings("ignore")

Table_Results_with_FIFA1 = Table_Results_with_FIFA

list_Fifa_Home_teams = Table_Results_with_FIFA1['Home_team_Fifa'].tolist()
list_Fifa_teams = Teams_Rating['Name'].tolist()

list_Fifa_Home_team_ATT = []

for n in list_Fifa_Home_teams:
    if n in list_Fifa_teams:
        Team_name_Fifa_ATT = Teams_Rating.loc[Teams_Rating['Name'] == n, 'ATT'].iloc[0]
        list_Fifa_Home_team_ATT.append(Team_name_Fifa_ATT)
    else:
        list_Fifa_Home_team_ATT.append(67)

Table_Results_with_FIFA1['Home_team_Fifa_ATT'] = list_Fifa_Home_team_ATT
Table_Results_with_FIFA1
```

```python
# Adding all the columns with the ratings


import warnings
warnings.filterwarnings("ignore")

list_Fifa_Home_teams = Table_Results_with_FIFA['Home_team_Fifa'].tolist()
list_Fifa_Away_teams = Table_Results_with_FIFA['Away_team_Fifa'].tolist()
list_Fifa_teams = Teams_Rating['Name'].tolist()




# Home teams


list_Fifa_Home_team_ATT = []

for n in list_Fifa_Home_teams:
    if n in list_Fifa_teams:
        Team_name_Fifa_ATT = Teams_Rating.loc[Teams_Rating['Name'] == n, 'ATT'].iloc[0]
        list_Fifa_Home_team_ATT.append(Team_name_Fifa_ATT)
    else:
        list_Fifa_Home_team_ATT.append(67)

Table_Results_with_FIFA['Home_team_Fifa_ATT'] = list_Fifa_Home_team_ATT




list_Fifa_Home_team_MID = []

for n in list_Fifa_Home_teams:
    if n in list_Fifa_teams:
        Team_name_Fifa_MID = Teams_Rating.loc[Teams_Rating['Name'] == n, 'MID'].iloc[0]
        list_Fifa_Home_team_MID.append(Team_name_Fifa_MID)
    else:
        list_Fifa_Home_team_MID.append(67)

Table_Results_with_FIFA['Home_team_Fifa_MID'] = list_Fifa_Home_team_MID




list_Fifa_Home_team_DEF = []

for n in list_Fifa_Home_teams:
    if n in list_Fifa_teams:
        Team_name_Fifa_DEF = Teams_Rating.loc[Teams_Rating['Name'] == n, 'DEF'].iloc[0]
        list_Fifa_Home_team_DEF.append(Team_name_Fifa_DEF)
    else:
        list_Fifa_Home_team_DEF.append(67)

Table_Results_with_FIFA['Home_team_Fifa_DEF'] = list_Fifa_Home_team_DEF




list_Fifa_Home_team_OVR = []

for n in list_Fifa_Home_teams:
    if n in list_Fifa_teams:
        Team_name_Fifa_OVR = Teams_Rating.loc[Teams_Rating['Name'] == n, 'OVR'].iloc[0]
        list_Fifa_Home_team_OVR.append(Team_name_Fifa_OVR)
    else:
        list_Fifa_Home_team_OVR.append(67)

Table_Results_with_FIFA['Home_team_Fifa_OVR'] = list_Fifa_Home_team_OVR








# Away_teams


list_Fifa_Away_team_ATT = []

for n in list_Fifa_Away_teams:
    if n in list_Fifa_teams:
        Team_name_Fifa_ATT = Teams_Rating.loc[Teams_Rating['Name'] == n, 'ATT'].iloc[0]
        list_Fifa_Away_team_ATT.append(Team_name_Fifa_ATT)
    else:
        list_Fifa_Away_team_ATT.append(67)

Table_Results_with_FIFA['Away_team_Fifa_ATT'] = list_Fifa_Away_team_ATT




list_Fifa_Away_team_MID = []

for n in list_Fifa_Away_teams:
    if n in list_Fifa_teams:
        Team_name_Fifa_MID = Teams_Rating.loc[Teams_Rating['Name'] == n, 'MID'].iloc[0]
        list_Fifa_Away_team_MID.append(Team_name_Fifa_MID)
    else:
        list_Fifa_Away_team_MID.append(67)

Table_Results_with_FIFA['Away_team_Fifa_MID'] = list_Fifa_Away_team_MID




list_Fifa_Away_team_DEF = []

for n in list_Fifa_Away_teams:
    if n in list_Fifa_teams:
        Team_name_Fifa_DEF = Teams_Rating.loc[Teams_Rating['Name'] == n, 'DEF'].iloc[0]
        list_Fifa_Away_team_DEF.append(Team_name_Fifa_DEF)
    else:
        list_Fifa_Away_team_DEF.append(67)

Table_Results_with_FIFA['Away_team_Fifa_DEF'] = list_Fifa_Away_team_DEF




list_Fifa_Away_team_OVR = []

for n in list_Fifa_Away_teams:
    if n in list_Fifa_teams:
        Team_name_Fifa_OVR = Teams_Rating.loc[Teams_Rating['Name'] == n, 'OVR'].iloc[0]
        list_Fifa_Away_team_OVR.append(Team_name_Fifa_OVR)
    else:
        list_Fifa_Away_team_OVR.append(67)

Table_Results_with_FIFA['Away_team_Fifa_OVR'] = list_Fifa_Away_team_OVR



# Re order columns

Table_Results_with_FIFA = Table_Results_with_FIFA[['Year', 'Weekday', 'Home_team', 'Away_team', 'Home_team_Fifa',
       'Away_team_Fifa', 'Result', 'Home_score', 'Away_score', '1x2',
       'Points_Home_Team', 'Points_Away_Team',  'Home_team_Fifa_OVR',
       'Away_team_Fifa_OVR', 'Home_team_Fifa_ATT', 'Home_team_Fifa_MID',
       'Home_team_Fifa_DEF', 'Away_team_Fifa_ATT', 'Away_team_Fifa_MID', 
       'Away_team_Fifa_DEF'
       ]]


Table_Results_with_FIFA
```
